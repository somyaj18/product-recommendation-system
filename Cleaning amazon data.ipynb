```python
import regex as re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def clean_text(text):
    if isinstance(text,str):
        text=text.lower()
        text=re.sub(r'\W',' ',text)
        text=re.sub(r'\s',' ',text)
        words=word_tokenize(text)
        words=[word for word in words if word not in stopwords.words('english')]
        return "  ".join(words)
    return ""
    
df_filtered["about_product_cleaned"]= df_filtered["about_product"].apply(clean_text)
df_filtered["review_content_cleaned"]= df_filtered["review_content"].apply(clean_text)
print(df_filtered.head())
```

```python
df_filtered.to_csv("filtered_amazon_data.csv",index=False)
```
